```python
# dropping columns
df.drop(columns = ['YearEnd', 'Datasource', 'Class', 'Topic','Data_Value_Unit', 'Data_Value_Type', 'DataValueTypeID',
                   'Question', 'Data_Value_Alt', 'Data_Value_Footnote', 'Data_Value_Footnote_Symbol', 'Total', 'Age(years)', 
                   'Gender', 'Income', 'Race/Ethnicity', 'StratificationCategory1', 'Stratification1', 'Education',
                   'LocationID', 'Low_Confidence_Limit', 'High_Confidence_Limit ', 'ClassID', 'TopicID'],
        inplace = True)
df
```

```python
# dropping US territories, sorry Guam
df.drop(df[df['LocationAbbr'] == 'GU'].index | df[df['LocationAbbr']  == 'PR'].index | df[df['LocationAbbr']  == 'VI'].index | df[df['LocationAbbr']  == 'US'].index,
        inplace = True)
# for some reason '|' operator works with warning, but 'or' doesn't
df
```

```python
# list of question id
qid_list = ['Q036', 'Q037', 'Q047', 'Q019', 'Q043', 'Q044', 'Q045', 'Q045', 'Q018']

# list of stratification categories (diff than stratification id)
strat_cat_list = ['RACE', 'INC', 'AGEYR', 'EDU', 'GEN', 'OVR']
```

```python
def strat_qid(strat_cat, qid):
    strat = df[df['StratificationCategoryId1'] == strat_cat][['YearStart', 'LocationDesc','Data_Value', 'QuestionID', 'StratificationID1']]
    question = strat[strat['QuestionID'] == qid][['YearStart', 'LocationDesc','Data_Value', 'QuestionID', 'StratificationID1']]
    return question
```

```python
strat_qid(strat_cat_list[0], qid_list[0]).tail(10)
```

```python
overall_strat = df[df['StratificationCategoryId1'] == strat_cat_list[0]][['YearStart', 'LocationDesc','Data_Value', 'QuestionID']]
# overall for % of adults aged 18+ who have obestiy
question = overall_strat[overall_strat['QuestionID'] == 'Q036'][['YearStart', 'LocationDesc','Data_Value', 'QuestionID']]
question
```

```python
# prints df with varying question id and stratification category parameters
for qid in qid_list:
    for strat in strat_cat_list:
        print(strat_qid(strat, qid))
```

```python
# only questions Q036, Q037, and Q047 have data for all years
# the remaining questions only have data every other year
df['QuestionID'].value_counts()
```
