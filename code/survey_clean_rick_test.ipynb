```python
import pandas as pd
```

```python
df = pd.read_csv('../data/survey.csv')
print(df.shape)
df.head(15)
```

```python
df.isnull().sum()
```

```python
df['family_history_with_overweight'] = df['family_history_with_overweight'].map({'yes':1, 'no':0})
df['FAVC'] = df['FAVC'].map({'yes':1, 'no':0})
df['SMOKE'] = df['SMOKE'].map({'yes':1, 'no':0})
df['SCC'] = df['SCC'].map({'yes':1, 'no':0})
```

```python
df = pd.get_dummies(data = df, columns = ['Gender', 'CAEC', 'CALC', 'MTRANS'], drop_first = True)
```

```python
df['NObeyesdad'].value_counts(normalize = True)
```

```python
#Change Values
df['NObeyesdad'] = df['NObeyesdad'].map({'Obesity_Type_I':3, 'Obesity_Type_II':3, 'Obesity_Type_III':3,
                                         'Overweight_Level_I':2, 'Overweight_Level_II':2, 'Normal_Weight':1, 'Insufficient_Weight':0})
```
